```python
# Problem 21: Given a sorted character array and a target letter, find the smallest letter in the array that is
# greater than the target.
# Input: letters = ['c', 'f', 'j'], target = a
# Output: 'c'


def nextGreatestAlphabet(ltr, K):
    l = len(ltr)
    if K>= ltr[l-1]:
        return "None"
    
    l = len(ltr)
    left = 0
    right = l-1
    while(left<=right):
        mid = int(left + (right - left)/2)
        if ltr[mid] > K:
            ans = mid
            right = mid - 1
        else:
            left = mid + 1
            
    if ltr[ans] < K:
        return ltr[0]
    else:
        return ltr[ans]
            
        
# Driver Code
letters = ['c', 'f', 'j']
K = 'a'

# Function call
result = nextGreatestAlphabet(letters, K)
print(result)
```

```python
# Problem 22: Given an array with n objects colored red, white, or blue, sort them in-place so that objects of
# the same color are adjacent, with the colors in the order red, white, and blue.
# Input: nums = [2, 0, 2, 1, 1, 0]
# Output: [0, 0, 1, 1, 2, 2]
def adjacent(arr):
    l = len(arr)
    a = 0
    mid = 0
    c = l -1
    while(mid <= c):
        if arr[mid] == 0:
            arr[a],arr[mid] = arr[mid],arr[a]
            mid = mid + 1
            a = a + 1
            
        elif arr[mid] == 1:
            mid = mid + 1
            
        else:
            arr[mid],arr[c] = arr[c],arr[mid]
            c = c - 1
    return arr
      
arr = [2,0,2,1,1,0]
adjacent(arr)
```

```python
# Problem 23: Find the kth largest element in an unsorted array.
# Input: nums = [3, 2, 1, 5, 6, 4], k = 2
# Output: 5

def k_largest(arr):
    l = len(arr)
    for i in range(0,l-1):
        a = i 
        for j in range(i+1,l):
            if arr[i] > arr[j]:
                i = j
        arr[a],arr[i] = arr[i],arr[a]   
    a = arr[l-2]
    return a
        

arr = [3,2,1,5,6,4]
k_largest(arr)
```

```python
#Problem 24: Given an unsorted array, reorder it in-place such that nums[0] <= nums[1] >= nums[2] <=
# nums[3]...
# Input: nums = [3, 5, 2, 1, 6, 4]
# Output: [3, 5, 1, 6, 2, 4]
def reorder(arr):
    l = len(arr)
    for i in range(0, len(arr)-1):
        if i % 2 == 0 :
            if arr[i] <= arr[i+1]:
                pass
            else:
                arr[i],arr[i+1] = arr[i+1],arr[i]
        if i % 2 != 0:
            if arr[i] >= arr[i+1]:
                pass
            else:
                arr[i],arr[i+1] = arr[i+1],arr[i]
    return arr
            
arr =  [3, 5, 2, 1, 6, 4]
reorder(arr)
```

```python
# Problem 25: Given an array of integers, calculate the sum of all its elements.
# Input: [1, 2, 3, 4, 5]
# Output: 15
def sum1(arr):
    sum12 = 0
    l = len(arr)
    for i in range(0, l):
        sum12 = arr[i] + sum12
    return sum12

arr = [ 1,2,3,4,5]
sum1(arr)      
```

```python
# Problem 26: Find the maximum element in an array of integers.
# Input: [3, 7, 2, 9, 4, 1]
# Output: 9
def largest(arr):
    num = 0
    l = len(arr)
    for i in range(0,l):
        if arr[i] > num:
            num = arr[i]
        
    return num     
    
arr = [3,7,2,9,4,1]
largest(arr)
```

```python
# Problem 27: Implement linear search to find the index of a target element in an array
# Input: [5, 3, 8, 2, 7, 4], target = 8
# Output: 2
def linear_search(arr,num):
    l = len(arr)
    for i in range(0,l):
        if arr[i] == num:
            return i 
arr = [5,3,8,2,7,4]
linear_search(arr,8)
```

```python
# Problem 28 Calculate the factorial of a given number.
# Input: 5
# Output: 120 (as 5! = 5 * 4 * 3 * 2 * 1 = 120)
def factorial(num):
    if num == 0:
        return 1
    a = num * factorial(num-1)
    return a
factorial(5)
```

```python
# Problem 29: Check if a given number is a prime number.
# Input: 7
# Output: True
def prime(num):
    for i in range(2,num-1):
        if num % i == 0:
            return False
        else:
            return True
num = 7
prime(num)
```

```python
# Problem 30: Generate the Fibonacci series up to a given number n.
# Input: 8
# Output: [0, 1, 1, 2, 3, 5, 8, 13]
def fibonacci(n):
    a = 0 
    b = 1
    print(a, end = ",")
    print(b, end = ",")
    for i in range(2,n):
        sum = a + b
        print(sum, end = ",")
        a = b
        b = sum
  
        
fibonacci(8)
```

```python
# Problem 31: Calculate the power of a number using recursion.
# Input: base = 3, exponent = 4
# Output: 81 (as 3^4 = 3 * 3 * 3 * 3 = 81)
def power(base, exponent):
    if exponent == 0:
        return 1
    
    result = power(base, exponent-1)
    return result * base
    
power(3,4)    
```
